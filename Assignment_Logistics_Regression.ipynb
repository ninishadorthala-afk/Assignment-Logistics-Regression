1. What is Simple Linear Regression (SLR)? Explain its purpose.
   - Simple Linear Regression is a method to predict one variable using another. It finds a straight-line relationship between them to make future predictions.


2. What are the key assumptions of Simple Linear Regression?
   - The assumptions are:
      - The relationship between variables is linear
      - Errors have constant variance
      - Errors are independent
      - Errors are normally distributed


3. Write the mathematical equation for a simple linear regression     model and explain each term.
  - The equation is y = mx + c
    y is the predicted value, x is the input, m is the slope, and c is the intercept.


4. Provide a real-world example where simple linear regression can be applied.
   - Predicting a personâ€™s weight based on their height is a common example of using simple linear regression.


5. What is the method of least squares in linear regression?
   - It is a technique to find the best-fitting line by minimizing the sum of squared differences between actual and predicted values.

6. What is Logistic Regression? How does it differ from Linear Regression?
    - Logistic Regression is used for classification. It predicts probabilities between 0 and 1, while Linear Regression predicts continuous values.


7. Name and briefly describe three common evaluation metrics for regression models.
      - Mean Absolute Error: average of absolute errors
      - Mean Squared Error: average of squared errors
      - R-squared: shows how well the model explains the data


8. What is the purpose of the R-squared metric in regression analysis?
    - R-squared tells how much of the variation in the output is explained by the input. Higher value means better fit.


```python
# 9. Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.

from sklearn.linear_model import LinearRegression
import numpy as np

x = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 5, 4, 5])

model = LinearRegression()
model.fit(x, y)

print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
```

10. How do you interpret the coefficients in a simple linear regression model?
    - The slope shows how much the output changes for each unit increase in input. The intercept is the output when input is zero.
